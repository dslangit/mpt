# SD and Return Calc

```python
# Arithmetic Return
def ari_ret(column):
    ari_ret = column/column.shift(1)-1
    return ari_ret
# Log Return
def log_ret(column):
    log_ret = np.log(column/column.shift(1))
    return log_ret
```

```python
# Return matrix for arithmetic
ari_rets = final_df.apply(ari_ret)
ari_rets
```

```python
#Correlation Matrix
final_df_corr = final_df.corr()
plt.figure(figsize = (10,6))
sns.heatmap(final_df_corr, vmin = 0., vmax = 1., cmap = sns.cubehelix_palette(as_cmap = True), annot = True)
plt.title('Correlation Matrix for the Crypto Assets', fontsize = 16) 
plt.savefig('corrmat.png')
plt.show()
```

```python
num_iter = 10000
port_weight = []
np.random.seed(42)
for i in range(num_iter):
    weight = np.random.random(len(final_df.columns))
    weight = weight/sum(weight)
    port_weight.append(weight)
print(port_weight)
print(len(port_weight))
```

```python
#Put the weight in the df
ari_df = pd.DataFrame(port_weight, columns = final_df.columns)
ari_df
```

# Calc Iter with Arithmetic Returns

## Return Calc

```python
# Mean Calc --> ari_base_rets
ari_df_result = ari_df.copy()
ari_base_rets = ari_rets.mean()
ari_base_rets
```

```python
ari_df_result['Port Return'] = np.dot(ari_df, ari_base_rets)
ari_df_result
```

## Covariance Calc

```python
#Covariance Matrix
ari_rets_cov = ari_rets.cov()
ari_rets_cov
```

```python
# covariance matrix = 13 x 13, weight matrix (ari_df) = 10,000 x 13 --> transposed 13 x 10,000, dot product = 13 x 10,000
portfolio_variances = np.sum(np.dot(ari_rets_cov, ari_df.T) * ari_df.T, axis = 0)
portfolio_variances
```

```python
ari_df_result['Port SD'] = np.sqrt(portfolio_variances)
ari_df_result #A table of weights, return, and volatility of each portfolio combination
```

# Optimal Portfolio

```python
sns.scatterplot(data = ari_df_result, x = 'Port SD', y = 'Port Return', color = 'blue')
plt.title('Portfolio: 10,000 Iterations')
plt.savefig('portfolio_blank.png')
plt.show()
```

## Max Sharpe Ratio

```python
#riskfree rate:
rf = 0.03 
```

```python
max_sr_port = ari_df_result.iloc[((ari_df_result['Port Return'] - rf)/ari_df_result['Port SD']).idxmax()]
max_sr_port
```

```python
max_sr_port_sum = max_sr_port[0:13].sum()
max_sr_port_sum
```

## Final Comparison with Baseline

```python
# Baseline Portfolio of equal weight

base_weight =  len(final_df.columns) * [(1/len(final_df.columns))]
base_ari_ret = np.dot(base_weight, ari_base_rets)
# base_ari_ret
base_ari_var = np.sum(np.dot(ari_rets_cov, np.array(base_weight).T) * np.array(base_weight).T, axis = 0)
base_ari_sd = np.sqrt(base_ari_var)
# base_ari_sd
base_weight.extend([base_ari_ret, base_ari_sd])
                                 
print(base_weight)
```

```python
# Table Comparison

max_sr_port = max_sr_port.to_frame().T
min_var_port = min_var_port.to_frame().T
compare_table = pd.concat([max_sr_port, min_var_port], axis = 0)
compare_table
```

```python
slope = (max_sr_port['Port Return'] - rf)/ max_sr_port['Port SD']
y_intercept = rf  

# Calculate y-values for the line using the slope and y-intercept
x_values = np.array([0, ari_df_result['Port SD'].max()])
y_values = slope * x_values + y_intercept

# Plot Max SR 
sr = (ari_df_result['Port Return'] - rf)/ari_df_result['Port SD']
idx_sr = ((ari_df_result['Port Return'] - rf)/ari_df_result['Port SD']).idxmax()
sr_risk = ari_df_result.loc[idx_sr, 'Port SD']
sr_return = ari_df_result.loc[idx_sr, 'Port Return']
plt.scatter(sr_risk, sr_return, c = 'blue', marker = '*', s = 100)  
# Adding text box for the annotated point
annotation_text = f'Max Sharpe Ratio\nReturn: {sr_return:.3f}\nRisk: {sr_risk:.3f}'
plt.text(sr_risk, sr_return + 0.003, annotation_text, color = 'blue', 
         verticalalignment = 'bottom', horizontalalignment = 'right')

# Plot Min Var 
idx_minvar = ari_df_result['Port SD'].idxmin()
minvar_risk = ari_df_result.loc[idx_minvar, 'Port SD']
minvar_return = ari_df_result.loc[idx_minvar, 'Port Return']
plt.scatter(minvar_risk, minvar_return, c = 'red', marker = '*', s = 100)  
# Adding text box for the annotated point
annotation_text = f'Min Variance\nReturn: {minvar_return:.3f}\nRisk: {minvar_risk:.3f}'
plt.text(minvar_risk, minvar_return + 0.003, annotation_text, color = 'black', 
         verticalalignment = 'bottom', horizontalalignment = 'right')

# Plot the line on the scatter plot
sns.scatterplot(data = ari_df_result, x = 'Port SD', y = 'Port Return', c = sr, cmap = 'plasma')
plt.colorbar(label = 'Sharpe Ratio')
plt.plot(x_values, y_values, color = 'red', label = 'Optimal Sharpe Ratio Line')
plt.title('Portfolio: 10,000 Iterations with Sharpe Ratio')
plt.legend()
plt.savefig('eff_frontier.png')
plt.show()
```
